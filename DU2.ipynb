## Jak to má fungovat

- Derivace je dobře algoritmizovatelná, protože se řídí prakticky pouze třemi pravidly (pravidlu podílu se lze vyhnout přes násobení funkcí $\frac{1}{f(x)}$):
    - derivace součtu je součet derivací,
    - pravidlo pro derivaci součinu,
    - derivace složené funkce je derivace vnější funkce násobená derivací argumentu.

- Stačí tedy implementovat tato tři základní pravidla a pro každou elementární funkci (`sin`, `cos`, `exp`, `ln`, $x^n$, ...) znát její derivaci.
- Celý proces lze reprezentovat pomocí tříd a dědičnosti.
- Trik je v tom, že se výrazy automaticky skládají do stromu (pro nás v podstatě skrytého), a tedy každý výraz je navenek pouze buď sčítání, násobení, nebo elementární funkce (a o jejich argumentech platí totéž).

## Podrobnější zadání
Konkrétně:
- Vytvořte třídu **Operator**.
    - Třída bude mít metody:
        - `__init__()` - inicializuje operátor (zde stačí `pass`),
        - `__str__()` - vrací řetězec reprezentující operátor,
        - `__repr__()` - vrací řetězec reprezentující operátor.
- Vytvořte třídu **BinaryOperator**, která bude dědit od třídy `Operator`.
    - Třída bude mít atributy:
        - `levy_argument` - levý operand,
        - `pravy_argument` - pravý operand.
    - Třída bude mít metody:
        - `__init__()` - inicializuje operátor, přiřadí argumenty,
        - `__str__()` - vrací řetězec reprezentující operátor,
        - `__repr__()` - vrací řetězec reprezentující operátor,
        - `diff()` - vrací derivaci operátoru.
- Vytvořte třídu **UnaryOperator**, která bude dědit od třídy `Operator`.
    - Třída bude mít atribut:
        - `argument` - argument funkce.
    - Třída bude mít metody:
        - `__init__()` - inicializuje operátor, přiřadí argument,
        - `__str__()` - vrací řetězec reprezentující operátor,
        - `__repr__()` - vrací řetězec reprezentující operátor,
        - `diff()` - vrací derivaci operátoru.

Pro počítání derivací použijte standardní pravidla:
- derivace součtu je součet derivací,
- pravidlo pro derivaci součinu,
- derivace složené funkce je derivace vnější funkce násobená derivací argumentu.

Pro **BinaryOperator** a **UnaryOperator** implementujte také metody standardních operátorů:
- `__add__()` - implementuje operátor `+`,
- `__sub__()` - implementuje operátor `-`,
- `__mul__()` - implementuje operátor `*`.

Pro **BinaryOperator** implementujte dvě dceřiné třídy:
- **Addition** - implementuje operátor `+`,
- **Multiplication** - implementuje operátor `*`.

Pro **UnaryOperator** implementujte tyto dceřiné třídy:
- **exp** - implementuje funkci `exp`,
- **sin** - implementuje funkci `sin`,
- **cos** - implementuje funkci `cos`,
- **ln** - implementuje funkci `ln`,
- **frc** - implementuje funkci `1/x`,
- **identity** - implementuje identitu `x`,
- **cst** - implementuje konstantu `c`,
- **mocnina** - implementuje mocninu `x^n`.

U všech funkcí implementujte adekvátně metody `diff()` a `__str__()`, aby se výrazy vypisovaly v přehledné podobě.

Otestujte řešení na následujících funkcích:
- $sin(x + cos(x*x)) + 1$
- $x^2 + 2x + 1 + 2x^2$
- $x^2 + 2x + exp(x)$


```python
# kostra k doplnění, místa označená TODO doplňte

class Operator:
    def __init__(self):
        pass
    
    def __str__(self):
        return type(self).__name__
    
    def __repr__(self):
        return self.__str__()
    
    def diff(self):
        pass
    
    def __add__(self, other):
        return Addition(self, other)

    def __sub__(self, other):
        return Addition(self, Multiplication(const(-1), other))
    
    def __mul__(self, other):
        return Multiplication(self, other)
    
class BinaryOperator(Operator):
    def __init__(self, levy_argument, pravy_argument):
        self.levy_argument = levy_argument
        self.pravy_argument = pravy_argument
        
class UnaryOperator(Operator):
    def __init__(self, argument):
        self.argument = argument
    
    def __str__(self):
        return type(self).__name__ + "(" + str(self.argument) + ")"
        
class Addition(BinaryOperator):
    def __str__(self):
        return f"{self.levy_argument} + {self.pravy_argument}"

    def diff(self):
        return Addition(self.levy_argument.diff(), self.pravy_argument.diff())
    
class Multiplication(BinaryOperator):
    def __str__(self):
        return f"({self.levy_argument})*({self.pravy_argument})"

    def diff(self):
        return Addition(Multiplication(self.levy_argument.diff(), self.pravy_argument), 
                        Multiplication(self.levy_argument, self.pravy_argument.diff()))
    
class const(UnaryOperator):
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return str(self.value)

    def diff(self):
        return const(0)
    
class identity(UnaryOperator):
    def __init__(self):
        pass

    def __str__(self):
        return "x"

    def diff(self):
        return const(1)
    
class mocnina(UnaryOperator):
    def __init__(self, argument, exponent):
        self.argument = argument
        self.exponent = exponent

    def __str__(self):
        return f"({self.argument})^{self.exponent}"
    
    def diff(self):
        return Multiplication(Multiplication(self.exponent, mocnina(self.argument, self.exponent - 1)), self.argument.diff())
    
class sin(UnaryOperator):
    def diff(self):
        return Multiplication(cos(self.argument), self.argument.diff())

class cos(UnaryOperator):
    def diff(self):
        return Multiplication(Multiplication(const(-1), sin(self.argument)), self.argument.diff())

class exp(UnaryOperator):
    def diff(self):
        return Multiplication(exp(self.argument), self.argument.diff())
    
class ln(UnaryOperator):
    def diff(self):
        return Multiplication(frc(self.argument), self.argument.diff())
    
class frc(UnaryOperator):
    def __str__(self):
        return f"1 / ({self.argument})"
    
    def diff(self):
        return Multiplication(const(-1), frc(Multiplication(mocnina(self.argument, 2), self.argument.diff())))
```

## Bonusový úkol
Doplňte ke všem třídám metodu `simplify()`, která provede zjednodušení výrazu. Například:
- součet s nulou vrátí druhý operand,
- součin s nulou vrátí nulu,
- mocnina s exponentem 0 vrátí 1,
- mocnina s exponentem 1 vrátí první argument,
- součin dvou stejných argumentů vrátí mocninu s exponentem 2,
- součin konstant vrátí konstantu se součinem,
- ...

Doplňte metody `__add__`, `__radd__`, `__mul__`, `__rmul__`, ... tak, aby podporovaly operace s čísly (např. `2 + x` nebo `x * 3`) a zjednodušovaly zadávání výrazů.

```python
# rozhodl jsem se trochu pozměnit funkce ať se s nima lépe pracuje
# například "Pow" je schopná zvládat i složitější případy jako x ^ x, 
# ale obecnější vzorečeck často vyhodí výsledek ve formě, která není tak jednoduchá k úpravě
# np x ^ 2 skončí jako x ^ 2 * 2 / x místo x * 2, protože simplify() neprohledává celý strom rekurzivně
# ještě jsem upravil vypisování závorek, ať je výsledný vzorec přehlednější

class Operator:
    def __init__(self):
        pass
    
    def __str__(self):
        return type(self).__name__
    
    def __repr__(self):
        return self.__str__()
    
    def diff(self):
        pass

    def simplify(self):
        return self.simplify()

    def __eq__(self, value):
        return False

    def __neg__(self):
        return Neg(self)
    
    def __add__(self, other):
        if isinstance(other, Operator):
            return Add(self, other)

        return Add(self, const(other))
    
    def __radd__(self, other):
        return const(other) + self

    def __sub__(self, other):
        if isinstance(other, Operator):
            return Sub(self, other)

        return Sub(self, const(other))
    
    def __rsub__(self, other):
        return const(other) - self

    def __mul__(self, other):
        if isinstance(other, Operator):
            return Mul(self, other)

        return Mul(self, const(other))
    
    def __rmul__(self, other):
        return const(other) * self
    
    def __truediv__(self, other):
        if isinstance(other, Operator):
            return Div(self, other)

        return Div(self, const(other))
    
    def __rtruediv__(self, other):
        return const(other) / self
    
    def __pow__(self, other):
        if isinstance(other, Operator):
            return Pow(self, other)

        return Pow(self, const(other))
    
    def __rpow__(self, other):
        return const(other) ** self
        
class BinaryOperator(Operator):
    def __init__(self, l_arg, r_arg):
        self.l_arg = l_arg
        self.r_arg = r_arg

    def __eq__(self, value):
        if type(self) != type(value):
            return False
        
        return self.l_arg == value.l_arg and self.r_arg == value.r_arg
        
class UnaryOperator(Operator):
    def __init__(self, arg):
        self.arg = arg
    
    def __str__(self):
        return type(self).__name__ + "(" + str(self.arg) + ")"
    
    def __eq__(self, value):
        if type(self) != type(value):
            return False
        
        return self.arg == value.arg
        
class Constant(Operator):
    def __init__(self):
        pass
    
    def __str__(self):
        pass
    
    def __eq__(self, value):
        if type(self) != type(value):
            return False
        
        if isinstance(self, const) and (self.arg != value.value):
            return False
        
        return True
    
    def simplify(self):
        return self
        
class identity(Constant):
    def __init__(self):
        pass

    def __str__(self):
        return "x"

    def diff(self):
        return const(1)
    
class euler(Constant):
    def __init__(self):
        pass

    def __str__(self):
        return "e"

    def diff(self):
        return const(0)
    
class const(Constant):
    def __init__(self, value):
        self.arg = value

    def __str__(self):
        return str(self.arg)

    def diff(self):
        return const(0)
  
class Add(BinaryOperator):
    def __str__(self):
        return f"{self.l_arg} + {self.r_arg}"

    def diff(self):
        return self.l_arg.diff() + self.r_arg.diff()
    
    def simplify(self):
        l = self.l_arg.simplify()
        r = self.r_arg.simplify()

        if isinstance(l, const) and isinstance(r, const):
            return const(l.arg + r.arg)
        
        if isinstance(l, const) and l.arg == 0:
            return r
        
        if isinstance(r, const) and r.arg == 0:
            return l
        
        if isinstance(l, Neg) and isinstance(r, Neg):
            return (- (l.arg + r.arg)).simplify()
        
        if isinstance(l, Neg) and (not isinstance(r, Neg)):
            return (r - l.arg).simplify()
        
        if (not isinstance(l, Neg)) and isinstance(r, Neg):
            return (l - r.arg).simplify()
        
        if l == r:
            return 2 * l
        
        return l + r
    
class Sub(BinaryOperator):
    def __str__(self):
        return f"{self.l_arg} - {self.r_arg}"

    def diff(self):
        return self.l_arg.diff() - self.r_arg.diff()
    
    def simplify(self):
        l = self.l_arg.simplify()
        r = self.r_arg.simplify()

        if isinstance(l, const) and isinstance(r, const):
            return const(l.arg - r.arg)
        
        if isinstance(l, const) and l.arg == 0:
            return -r
        
        if isinstance(r, const) and r.arg == 0:
            return l
        
        if isinstance(l, Neg) and isinstance(r, Neg):
            return (r.arg - l.arg).simplify
        
        if isinstance(l, Neg) and (not isinstance(r, Neg)):
            return (- (l.arg + r)).simplify()
        
        if (not isinstance(l, Neg)) and isinstance(r, Neg):
            return (l + r.arg).simplify()
        
        if l == r:
            return const(0)
        
        return l - r

class Mul(BinaryOperator):
    def __str__(self):
        l_str = f"({self.l_arg})" if isinstance(self.l_arg, (Add, Sub, Neg)) else f"{self.l_arg}"
        r_str = f"({self.r_arg})" if isinstance(self.r_arg, (Add, Sub, Neg)) else f"{self.r_arg}"

        return f"{l_str} * {r_str}"

    def diff(self):
        return self.l_arg.diff() * self.r_arg + self.l_arg * self.r_arg.diff()
    
    def simplify(self):
        l = self.l_arg.simplify()
        r = self.r_arg.simplify()

        if isinstance(l, const) and isinstance(r, const):
            return const(l.arg * r.arg)
        
        if isinstance(l, const) and l.arg == 0:
            return const(0)
        
        if isinstance(r, const) and r.arg == 0:
            return const(0)
        
        if isinstance(l, const) and l.arg == 1:
            return r
        
        if isinstance(r, const) and r.arg == 1:
            return l
        
        if isinstance(l, Neg) and isinstance(r, Neg):
            return (l.arg * r.arg).simplify()
        
        if isinstance(l, Neg) != isinstance(r, Neg):
            return (- l.arg * r.arg).simplify()
        
        if isinstance(l, Pow) and isinstance(r, Pow) and l.l_arg == r.l_arg:
            return (l.l_arg ** (l.r_arg + r.r_arg)).simplify()
        
        if isinstance(l, Pow) and l.l_arg == r:
            return (r ** (l.r_arg + 1)).simplify()
        
        if isinstance(r, Pow) and l == r.l_arg:
            return (l ** (r.r_arg + 1)).simplify()
        
        if isinstance(l, Div) and isinstance(r, Div):
            return ((l.l_arg * r.l_arg) / (l.r_arg * r.r_arg)).simplify()
        
        if isinstance(l, Div):
            return ((l.l_arg * r) / l.r_arg).simplify()
        
        if isinstance(r, Div):
            return ((l* r.l_arg) / r.r_arg).simplify()
        
        if l == r:
            return l ** 2
        
        return l * r
    
class Div(BinaryOperator):
    def __str__(self):
        l_str = f"({self.l_arg})" if isinstance(self.l_arg, (Add, Sub, Neg)) else f"{self.l_arg}"
        r_str = f"({self.r_arg})" if isinstance(self.r_arg, (Add, Sub, Neg)) else f"{self.r_arg}"

        return f"{l_str} / {r_str}"

    def diff(self):
        return self.l_arg.diff() * self.r_arg - self.l_arg * self.r_arg.diff() / self.r_arg ** 2
    
    def simplify(self):
        l = self.l_arg.simplify()
        r = self.r_arg.simplify()

        if isinstance(l, const) and isinstance(r, const):
            return const(l.arg / r.arg)
        
        if isinstance(l, const) and l.arg == 0:
            return const(0)
        
        if isinstance(r, const) and r.arg == 0:
            raise Exception("Cant divide by 0")
        
        if isinstance(r, const) and r.arg == 1:
            return l
        
        if isinstance(l, Neg) and isinstance(r, Neg):
            return (l.arg * r.arg).simplify()
        
        if isinstance(l, Neg) != isinstance(r, Neg):
            return (- l.arg * r.arg).simplify()
        
        if l == r:
            return const(1)
        
        return l / r

class Pow(BinaryOperator):
    def __str__(self):
        l_str = f"({self.l_arg})" if isinstance(self.l_arg, (BinaryOperator, Neg)) else f"{self.l_arg}"
        r_str = f"({self.r_arg})" if isinstance(self.r_arg, (BinaryOperator, Neg)) else f"{self.r_arg}"

        return f"{l_str} ^ {r_str}"

    def diff(self):
        return self * (ln(self.l_arg) * self.r_arg).diff()
    
    def simplify(self):
        l = self.l_arg.simplify()
        r = self.r_arg.simplify()

        if isinstance(l, const) and isinstance(r, const):
            return const(l.arg ** r.arg)
        
        if isinstance(l, const) and l.arg == 0:
            return const(0)
        
        if isinstance(r, const) and r.arg == 0:
            return const(1)
        
        if isinstance(l, const) and l.arg == 1:
            return const(1)
        
        if isinstance(r, const) and r.arg == 1:
            return l
        
        if isinstance(l, Neg) and isinstance(r, const) and r.arg % 2 == 0:
            return l.arg ** r
        
        if isinstance(l, Neg) and isinstance(r, const) and r.arg % 2 == 1:
            return - l.arg ** r
        
        return l ** r

class Neg(UnaryOperator):
    def __str__(self):
        if isinstance(self.arg, Add):
            return f"-({self.arg})"
        return f"-{self.arg}"

    def diff(self):
        return -self.arg.diff()
    
    def simplify(self):
        a = self.arg.simplify()

        if isinstance(a, const):
            return const(-a.arg)
        
        if isinstance(a, Neg):
            return a.value
        
        return -a
    
class sin(UnaryOperator):
    def diff(self):
        return cos(self.arg) * self.arg.diff()
    
    def simplify(self):
        a = self.arg.simplify()

        return sin(a)

class cos(UnaryOperator):
    def diff(self):
        return -sin(self.arg) * self.arg.diff()
    
    def simplify(self):
        a = self.arg.simplify()

        return cos(a)
    
class ln(UnaryOperator):
    def diff(self):
        return self.arg.diff() / self.arg

    def simplify(self):
        a = self.arg.simplify()

        if isinstance(a, euler):
            return const(1)
        
        return ln(a)
    
```

```python
x = identity()
e = euler()

f = sin(x ** x) * ln(e) + e ** x + 20

print(f)
print(f.diff().simplify())
```
